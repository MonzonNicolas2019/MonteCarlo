$$\int_0^1 \! x^2 \, dx = \frac{1}{3}$$

```python
def graficar_approx_montecarlo(f, nmax, a, b):
    inside = 0

    x_inside = []
    y_inside =  []
    x_outside = []
    y_outside = []

    for _ in range(nmax):
        x_random = random() * a
        y_random = random() * b
        y = f(x_random)
        if y_random <= y:
            inside += 1
            x_inside.append(x_random)
            y_inside.append(y_random)
        else:
            x_outside.append(x_random)
            y_outside.append(y_random)

    totalArea = inside/nmax #Porcentaje de figura respecto a la caja

    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,10))
    ax.set_aspect('equal')
    ax.scatter(x_inside, y_inside, color='g', marker='s')
    ax.scatter(x_outside, y_outside, color='r', marker='s')
    z = np.linspace(0,a,1000)
    ax.plot(z,f(z))
    fig.tight_layout()
    fig.show()
    print("Area aproximada: ", a*b*totalArea)
    print("Area real: ", quad(f,0,a))

def preparar_approx_montecarlo(ftext, nmax, a, b):
    fcode = parser.expr(ftext).compile()
    f = lambda t: eval(fcode)
    graficar_approx_montecarlo(f, nmax, a, b)


```

```python
def graficar_approx_montecarlo_and_calculate_pi():
    
    inside = 0
    n = 10**3
    x_inside = []
    y_inside =  []
    x_outside = []
    y_outside = []

    for _ in range(n):
        x = random()*2
        y = random()*2
        if (x-1)**2+(y-1)**2 <= 1:
            inside += 1
            x_inside.append(x)
            y_inside.append(y)
        else:
            x_outside.append(x)
            y_outside.append(y)

    totalArea = inside/n #Porcentaje de figura respecto a la caja
    pi = 4*totalArea
    
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,10))
    ax.set_aspect('equal')
    ax.scatter(x_inside, y_inside, color='g', marker='s')
    ax.scatter(x_outside, y_outside, color='r', marker='s')
    
    fcode = parser.expr('sqrt(1-(t-1)**2)+1').compile()
    f = lambda t: eval(fcode)
    z = np.linspace(0,2,1000)
    ax.plot(z,f(z))
    
    fcode = parser.expr('-sqrt(1-(t-1)**2)+1').compile()
    f = lambda t: eval(fcode)
    z = np.linspace(0,2,1000)
    ax.plot(z,f(z))
    
    fig.tight_layout()
    fig.show()
    print("Pi aprox: ", pi)
```

```python
graficar_approx_montecarlo_and_calculate_pi()
```
